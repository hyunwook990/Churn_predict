```python
import os
import sys
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
from datetime import datetime, timezone
import pytz
```

```python
path = os.path.join(os.path.split(os.getcwd())[0], "data\\reduction_data\\*.csv")
file_path = glob(path)

player_info_data = pd.DataFrame()

for file in file_path:
    player_info_data = pd.concat([player_info_data, pd.read_csv(file, encoding="utf-8")], axis=0)

display(player_info_data.head())
print(f"Player information data count : {len(player_info_data.index)}")
```

```python
path = os.path.join(os.path.split(os.getcwd())[0], "data\\summoner\\*.csv")
file_path = glob(path)

summoner_data = pd.DataFrame()

for file in file_path:
    summoner_data = pd.concat([summoner_data, pd.read_csv(file, encoding="utf-8")], axis=0)

display(summoner_data.head())
print(f"Summoner data count : {len(summoner_data.index)}")
```

```python
data = pd.merge(player_info_data, summoner_data, how="outer", on="puuid")
data.head()
```

```python
def convert_utc_to_kst(utc_timestamp):
    """UTC timestamp (밀리초)를 KST (한국 시간)으로 변환"""
    
    if isinstance(utc_timestamp, datetime):
        utc_time = utc_timestamp
    else:
        utc_time = datetime.fromtimestamp(utc_timestamp / 1000, tz=datetime.timezone.utc)
    
    kst_zone = pytz.timezone("Asia/Seoul")
    kst_time = utc_time.astimezone(kst_zone)
    return kst_time

def time_since(rev_date):
    """현재 시간과 주어진 시간(kst_time) 사이의 경과 시간을 반환"""
    now = datetime.now(pytz.timezone("Asia/Seoul"))  # 현재 한국 시간
    delta = now - rev_date

    days = delta.days
    years, remainder_days = divmod(days, 365)
    months, remainder_days = divmod(remainder_days, 30)
    hours, remainder = divmod(delta.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)

    if years > 0:
        return f"{years}년 전"
    elif months > 0:
        return f"{months}개월 전"
    elif days > 0:
        return f"{days}일 전"
    elif hours > 0:
        return f"{hours}시간 전"
    elif minutes > 0:
        return f"{minutes}분 전"
    else:
        return f"{seconds}초 전"
    
def days_since(rev_date):
    """현재 시간과 주어진 시간(kst_time) 사이의 경과 시간을 반환"""
    now = datetime.now(pytz.timezone("Asia/Seoul"))  # 현재 한국 시간
    delta = now - rev_date

    days = delta.days

    return days
```

```python
since_last_rev = []
days_rev = []
churn = []
df = data.dropna()

for i in df['revisionDate']:
        rev_date = convert_utc_to_kst(datetime.fromtimestamp(i / 1000, tz=timezone.utc))
        since_last_rev.append(time_since(rev_date))
        days_rev.append(days_since(rev_date))
        churn.append(1 if (days_since(rev_date) >= 29) else 0)
        

since_last_rev

df["since_rev"] = since_last_rev
df["days_since_last_rev"] = days_rev
df["churn"] = churn
df = df.drop_duplicates()
df = df.reset_index()
df
```
