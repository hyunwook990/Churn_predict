```python
import os
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
path = os.path.join(os.path.split(os.getcwd())[0], "data\\EDA\\unoversampling_EDA.csv")

prep_data = pd.read_csv(path)
prep_data
```

```python
df = prep_data.drop(columns=['Unnamed: 0', "tier", "rank", "winRate", "veteran", "hotStreak", "revisionDate", "championId", "championLevel", "championPoints"])
df
```

```python
X = df.drop('churn', axis=1)
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
knn = KNeighborsClassifier()
lr = LogisticRegression()
dt = DecisionTreeClassifier()

model1 = VotingClassifier(
    estimators=[
        ('knn', knn),
        ('lr', lr),
        ('dt', dt)
    ],
    voting='hard'
)

model1.fit(X_train, y_train)

# 예측 및 평가
y_pred_train = model1.predict(X_train)
acc_score_train = accuracy_score(y_train, y_pred_train)
f1_train = f1_score(y_train, y_pred_train)
print("훈련 데이터 평가: ", acc_score_train)
print("훈련 데이터 f1 평가: ", f1_train)

y_pred_test = model1.predict(X_test)
acc_score_test = accuracy_score(y_test, y_pred_test)
f1_test = f1_score(y_test, y_pred_test)
print("평가 데이터 평가: ", acc_score_test)
print("평가 데이터 f1 평가: ", f1_test)
```

```python
df2 = prep_data.drop(columns=['Unnamed: 0',  "revisionDate", "championId"])
df2
```

```python
X = df2.drop('churn', axis=1)
y = df2['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
columns = ["churn"] # "veteran", "freshBlood", "churn"
oversampl_X_data = df2.copy()
smote = SMOTE(random_state=42)

for column in columns:
    oversampl_y_data = oversampl_X_data[column]
    oversampl_X_data = oversampl_X_data.drop(labels=column, axis=1)
    
    oversampl_X_data, oversampl_y_data = smote.fit_resample(oversampl_X_data, oversampl_y_data)
    oversampl_X_data = pd.concat([oversampl_X_data, oversampl_y_data], axis=1)

oversampl_data = oversampl_X_data
```

```python
oversampl_data
```

```python
X = oversampl_data.drop('churn', axis=1)
y = oversampl_data['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
knn = KNeighborsClassifier()
lr = LogisticRegression(max_iter=300)
dt = DecisionTreeClassifier()

model3 = VotingClassifier(
    estimators=[
        ('knn', knn),
        ('lr', lr),
        ('dt', dt)
    ],
    voting='hard'
)

model3.fit(X_train, y_train)

# 예측 및 평가
y_pred_train = model3.predict(X_train)
acc_score_train = accuracy_score(y_train, y_pred_train)
f1_train = f1_score(y_train, y_pred_train)

y_pred_test = model3.predict(X_test)
acc_score_test = accuracy_score(y_test, y_pred_test)
f1_test = f1_score(y_test, y_pred_test)
print("훈련 데이터 평가: ", acc_score_train)
print("평가 데이터 평가: ", acc_score_test)
print("훈련 데이터 f1 평가: ", f1_train)
print("평가 데이터 f1 평가: ", f1_test)
```

```python
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=200, learning_rate=0.03, random_state=42)
catboost = CatBoostClassifier(iterations=300, learning_rate=0.03, depth=10, verbose=0)  
rf = RandomForestClassifier(n_estimators=200, random_state=42)  
lgbm = LGBMClassifier(n_estimators=200, learning_rate=0.03, random_state=42) 


# VotingClassifier
model4 = VotingClassifier(
    estimators=[
        ('catboost', catboost),
        ('rf', rf),
        ('lgbm', lgbm),
        ('gb', gb)
    ],
    voting='soft'
)

model4.fit(X_train, y_train)

y_pred_train = model4.predict(X_train)
acc_score_train = accuracy_score(y_train, y_pred_train)
f1_train = f1_score(y_train, y_pred_train)

y_pred_test = model4.predict(X_test)
acc_score_test = accuracy_score(y_test, y_pred_test)
f1_test = f1_score(y_test, y_pred_test)

print("훈련 데이터 정확도: ", acc_score_train)
print("테스트 데이터 정확도: ", acc_score_test)
print("훈련 데이터 F1 스코어: ", f1_train)
print("테스트 데이터 F1 스코어: ", f1_test)
```
