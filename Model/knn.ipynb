```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score
```

```python
path = os.path.join(os.path.split(os.getcwd())[0], "data\\EDA\\EDA.csv")

data = pd.read_csv(path, encoding="utf-8")
data = data.drop(labels="revisionDate", axis=1)
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], train_size=0.75)
print(f"X train data shape : {X_train.shape}")
print(f"y train data shape : {y_train.shape}")
print(f"X test data shape : {X_test.shape}")
print(f"y test data shape : {y_test.shape}")
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)

print(f"훈련 데이터 평가 : {accuracy_score(y_train, y_train_pred)}")
print(f"평가 데이터 평가 : {accuracy_score(y_test, y_test_pred)}")
print(f"훈련 데이터 f1 평가 : {f1_score(y_test, y_test_pred)}")
print(f"평가 데이터 f1 평가 : {f1_score(y_test, y_test_pred)}")
```
