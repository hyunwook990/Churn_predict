```python
import os
import sys
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as player_list
```

### 데이터셋 가져오기

```python
with open("data/champion.json", "r") as j:
    champion_json = json.load(j)
champion = pd.DataFrame(champion_json['data'].values())

display(champion.head())
```

### Champion 데이터 정제

```python
# 데이터가 dict인 경우 이 데이터들로 새로운 df 생성
def get_champion_data(df, col):
    df_info = {key : [] for df_value in df[col].values for key in df_value.keys()}

    for value in df[col].values:
        for key, info in value.items():
            df_info[key].append(info)

    df_info = pd.DataFrame(df_info)

    return df_info

# 데이터가 list인 경우 list 풀어주기어주기
def list_data_change(df, col):
    data = {col : []}
    for tag in df[col]:
        data[col].append(', '.join(tag))
    
    data = pd.DataFrame(data)

    return data
```

```python
champion_info = get_champion_data(champion, "info")
champion_stats = get_champion_data(champion, "stats")
champion_tags = list_data_change(champion, "tags")

display(champion.head())
display(champion_info.head())
display(champion_stats.head())
display(champion_tags.head())
```

```python
# data row 갯수 확인
print(f"Champion count : {len(champion)}")
print(f"Champion information count : {len(champion_info)}")
print(f"Champion status count : {len(champion_stats)}")
print(f"Champion tag count : {len(champion_tags)}")
```

```python
champion = champion.drop(['version', 'id', 'title', 'blurb', 'info', 'image', 'tags', 'stats'], axis=1)
champion_df = pd.concat([champion, champion_info, champion_stats, champion_tags], axis=1)
champion_df
```

```python
champion_df.to_csv("data/champion_df.csv", index=False, encoding="utf-8")
```
