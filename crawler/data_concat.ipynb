```python
import os
import pandas as pd
from glob import glob
from collections import defaultdict
```

```python
player_info_data_file_path = os.path.join(os.getcwd(), "data\\reduction_data\\*.csv")
summoner_data_file_path = os.path.join(os.getcwd(), "data\\summoner\\*.csv")
player_info_data_file_path_list = glob(player_info_data_file_path)
summoner_data_file_path_list = glob(summoner_data_file_path)
print(player_info_data_file_path_list)
print(summoner_data_file_path_list)


player_info_data = pd.DataFrame()
summoner_data = pd.DataFrame()

for file_path in player_info_data_file_path_list:
    player_info_data = pd.concat([player_info_data, pd.read_csv(file_path, encoding="utf-8")], axis=0)

for file_path in summoner_data_file_path_list:
    summoner_data = pd.concat([summoner_data, pd.read_csv(file_path, encoding="utf-8")], axis=0)
```

```python
print("Player Information Data\n")
display(player_info_data.head())
print(f"\nPlayer information data count : {len(player_info_data.index)}\n")
print("Summoner Data\n")
display(summoner_data.head())
print(f"\nsummoner data count : {len(summoner_data.index)}\n")
```
