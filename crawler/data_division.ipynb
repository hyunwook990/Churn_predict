```python
import os
import pandas as pd
import numpy as np
from glob import glob
from collections import defaultdict
from itertools import product
from sklearn.model_selection import train_test_split
```

```python
path = os.path.join(os.getcwd(), "data\\player_info\\*_player_info.csv")
file_path_list = glob(path)

data = pd.DataFrame()

for file_path in file_path_list:
    data = pd.concat([data, pd.read_csv(file_path, index_col=0, encoding="utf-8")], axis=0)
```

```python
display(data.head(2))
display(data.tail(2))
print(f"Data count : {len(data.index)}")
```

```python
low_tiers = ["IRON", "BRONZE", "SILVER", "GOLD", "PLATINUM", "EMERALD", "DIAMOND"]
low_divisions = ["I", "II", "III", "IV"]
high_tiers = ["MASTER", "GRANDMASTER", "CHALLENGER"]
high_divisions = ["I"]

low_tiers_divisions = list(product(*[low_tiers, low_divisions]))
high_tiers_divisions = list(product(*[high_tiers, high_divisions]))
tiers_divisions = low_tiers_divisions + high_tiers_divisions

sample_data = defaultdict(pd.DataFrame)

for tier, division in tiers_divisions:
    sample_data[tier+division] = data[(data["tier"] == tier) & (data["rank"] == division)]
```

```python
data_reduced = defaultdict(pd.DataFrame)
for key, value in sample_data.items():
    data_reduced[key] = value.sample(frac=0.1)
    print(f"Original {key} data count : {len(sample_data[key].index)}")
    print(f"reduced {key} data count : {len(data_reduced[key].index)}")
```

```python
data_concat = pd.DataFrame()

for df in data_reduced.values():
    data_concat = pd.concat([data_concat, df], axis=0)
```
